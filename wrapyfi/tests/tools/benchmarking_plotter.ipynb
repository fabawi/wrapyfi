```python
init_notebook_mode(connected=True)         # initiate notebook for offline plot

df1_sub = pd.read_csv("results/benchmarking_native_object_listen_ros2.csv", delimiter=",")
df2_sub = pd.read_csv("results/benchmarking_native_object_listen_ros.csv", delimiter=",")
df3_sub = pd.read_csv("results/benchmarking_native_object_listen_yarp.csv", delimiter=",")
df4_sub = pd.read_csv("results/benchmarking_native_object_listen_zeromq.csv", delimiter=",")

df_sub = pd.concat([df1_sub,df2_sub, df3_sub, df4_sub])
df_sub["type"] = "subscriber"

df1_pub = pd.read_csv("results/benchmarking_native_object_publish_ros2.csv", delimiter=",")
df2_pub = pd.read_csv("results/benchmarking_native_object_publish_ros.csv", delimiter=",")
df3_pub = pd.read_csv("results/benchmarking_native_object_publish_yarp.csv", delimiter=",")
df4_pub = pd.read_csv("results/benchmarking_native_object_publish_zeromq.csv", delimiter=",")
df_pub = pd.concat([df1_pub,df2_pub, df3_pub, df4_pub])
df_pub["type"] = "publisher"

df = pd.concat([df_sub,df_pub])
df['middleware'] = df['middleware'].str.replace('ros2','ROS 2')
df['middleware'] = df['middleware'].str.replace('ros','ROS')
df['middleware'] = df['middleware'].str.replace('yarp','YARP')
df['middleware'] = df['middleware'].str.replace('zeromq','ZeroMQ')

#figure = px.line(df_sub,  x="count", y="delay", color="middleware", template="simple_white")
figure = px.box(df,  x="middleware", y="delay", color="type", points=False, log_y=True, template="simple_white") # points="all"
# trace1 = Scatter(
#   x=[1, 2, 3, 4],
#   y=[16, 5, 11, 9]
# )

# iplot([trace0])  
figure.show()
figure.write_image("results/native_object_benchmarks.pdf")
figure.write_image("results/native_object_benchmarks.pdf")
```
